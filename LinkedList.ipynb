```python
#include<stdio.h>
#include<stdlib.h>

struct node
{
	int data;
	struct node *next;
};

struct node* create()
{
    struct node *ptr = (struct node*)malloc(sizeof(struct node));
    return ptr;
}

void traverse(struct node *ptr)
{
	printf("\nLinkedList : \n");
	while(ptr!=NULL)
	{
		printf("%d\n",ptr->data);
		ptr=ptr->next;
	}
}

struct node * insert(struct node *head,int data)
{
	struct node* ptr = create();
	int c,x;

	printf("\nInsert at the 1.beginning ,2. in between,3. end,\n");
	scanf("%d",&c);

	switch(c)
	{
		case 1:{
				ptr->data=data;
				ptr->next =head;
				head=ptr;
				break;
			}

		
		case 2: {
				struct node *p=head;
				printf("\nEnter the position of the node to be inserted\n");
				scanf("%d",&x);
				
				for(int i=0;i<x-2;i++)
				{
					p=p->next;
				}
				
				ptr->data=data;
				ptr->next=p->next;
				p->next=ptr;
				break;
			}
		case 3: {
				ptr->data=data;
				struct node *p=head;
	
				while(p->next!=NULL)
				{
					p=p->next;
				}
				p->next=ptr;
				ptr->next=NULL;
				break;
			}
		default : printf("\nChoose from 1 to 3\n");
	}
	return head;
}

struct node * delete(struct node *head)
{
	int c,x;

	printf("\nDelete from 1.beginning ,2. in between,3. end\n");
	scanf("%d",&c);

	switch(c)
	{
		case 1: {
				struct node *ptr=head;
				head=head->next;
				free(ptr);
				break;
			}
		case 2:{
				struct node *p=head;
				struct node *q=head->next;
	
				printf("\nEnter the position of the node to be deleted\n");
				scanf("%d",&x);

				for(int i=0;i<x-2;i++)
				{
					p=p->next;
					q=q->next;
				}
				p->next=q->next;
				free(q);
				break;
			}
		case 3:{
				struct node *p=head;
				struct node *q=head->next;
	
				while(q->next!=NULL)
				{
					p=p->next;
					q=q->next;
				}
				p->next=NULL;
				free(q);
				break;
			}
		case 4: printf("\n Choose from 1 to 3\n");
	}
	return head;
}

void main()
{
	int data,k;
	struct node *head=NULL;
	do
	{	
		printf("\n 1. to insert, 2. to delete, 3. to display,4. to exit \n");
		scanf("%d",&k);
		switch(k)
		{
			case 1:{
					printf("\nEnter element to be inserted\n");
					scanf("%d",&data);

					head=insert(head,data);
					traverse(head);
					break;
				}
			case 2:{
					head=delete(head);
					traverse(head);
					break;
				}
			case 3: traverse(head);
				break;
			case 4: exit(0);
			default:printf("\nChoose from 1 to 4\n");
		}
	}while(1);
}
```
